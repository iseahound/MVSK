```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.sandbox.distributions.extras as extras

# Define a linear space, meaning a set of intervals on a line.
# Here we have the starting point at -10 up to the ending point at +10. 
# This creates 100 points, NOT 100 intervals.
x_axis = np.linspace(-10, 10, 100)

# Control - Gaussian Distribution @ mean = 0, standard deviation = 3.25
plt.plot(x_axis, scipy.stats.norm.pdf(x_axis, loc=0, scale=3.25), color='red', label='Normal Distribution')

# MVSK - Mean, VARIANCE, Skew, Kurtosis - Note that the stddev SQUARED is the variance.
plt.plot(x_axis, extras.pdf_mvsk([0, 10, 0, 0])(x_axis), label='MVSK (0, 10, 0, 0)')

# Show a fat tail. # Guide by iseahound
plt.plot(x_axis, extras.pdf_mvsk([0, 10, 0, 2.5])(x_axis), label='Kurtosis (0, 10, 0, 2.5)', color='hotpink')

# Show skew.
plt.plot(x_axis, extras.pdf_mvsk([0, 10, 3, 0])(x_axis), label='Skewed (0, 10, 3, 0)', color='mediumseagreen')

plt.legend()
plt.show()
```
